```python
import random

staying_wins = 0
switching_wins = 0
```

```python
for _ in range(10000000):
    # Car is behind one door, other two have goats
    car_door = random.randint(1, 4)
    
    # Contestant makes a random choice
    original_choice = random.randint(1, 4)
    
    # Monty opens a door out of the other two which definitely does not have the car
    if car_door == original_choice:
        # can open any of the other 2 doors
        if car_door == 1:
            opened_door = random.randint(2, 4)
        elif car_door == 3:
            opened_door = random.randint(1, 3)
        else:
            temp = random.randint(1, 3)
            if temp == 1:
                opened_door = 1
            else:
                opened_door = 3
    else:
        # must open the goat door
        if original_choice == 1:
            if car_door == 2:
                opened_door = 3
            else:
                opened_door = 2
        elif original_choice == 2:
            if car_door == 1:
                opened_door = 3
            else:
                opened_door = 1
        else:
            if car_door == 1:
                opened_door = 2
            else:
                opened_door = 1
    
    if original_choice == car_door:
        # Count a win for staying
        staying_wins += 1
    else:
        # Count win for switching
        switching_wins += 1
```

```python
print('stay wins = {}, switch wins = {}'.format(staying_wins, switching_wins))
```

```python
print('You would have won {}% of the time by switching.'.format((switching_wins/(switching_wins+staying_wins))*100))
```
